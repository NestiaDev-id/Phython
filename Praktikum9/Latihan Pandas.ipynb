```python
# Mengubah tipe kolom tanggal datang dan pulang ke pd datetime
df["Tanggal datang kunjungan FKTP"] = pd.to_datetime(df["Tanggal datang kunjungan FKTP"])
df["Tanggal pulang kunjungan FKTP"] = pd.to_datetime(df["Tanggal pulang kunjungan FKTP"])

df["Tahun datang"] = df["Tanggal datang kunjungan FKTP"].dt.year
df["Bulan datang"] = df["Tanggal datang kunjungan FKTP"].dt.month
df["Tanggal datang"] = df["Tanggal datang kunjungan FKTP"].dt.day

df["Tahun pulang"] = df["Tanggal pulang kunjungan FKTP"].dt.year
df["Bulan pulang"] = df["Tanggal pulang kunjungan FKTP"].dt.month
df["Tanggal pulang"] = df["Tanggal pulang kunjungan FKTP"].dt.day

df
```

```python
# Mengganti kolom kelas str menjadi numeric
df['Status'] = df ["Kelas Status Pulang Peserta"].map({"Sehat" : 1, "Belum_Sehat":0})
df
```

```python
# Fungsi missing check
def missing_check(df) :
    missing = df.isnull().sum()
    per_miss = missing/len(df)
    data_type = df.dtypes
    number_unique = df.nunique()
    return pd.DataFrame({"Missing":missing,
                        "Percent_Missing":per_miss,
                       "Data_Types": data_type,
                        "Number_of_unique":number_unique}).style.format({'Percent_Missing' : "{:.2%}"})
```

```python
missing_check(df)
```
